```python
# Import all dependencies
import numpy as np
import pandas as pd
import glob
from pandas import Series, DataFrame
from datetime import datetime, timedelta
```

```python
# Find all csv files in 2019
glob.glob('../data/JC-*.csv')
```

```python
# Create a loop and append to list of data for 2019-2020
list_data = []
for filename_19_20 in glob.glob('../data/JC-*.csv'):
#     print(f'Loading {filename_2019}')
    temp_data = pd.read_csv(filename_19_20)
    list_data.append(temp_data)
```

```python
# Check length of list has all csv files from 2019-2020
len(list_data)
```

```python
# Print list for 2019-2020
list_data
```

```python
# Concat all 2019-2020 csv files and place into final DataFrame for 2019-2020
final_bike_data_df = pd.concat([pd.read_csv(filename_19_20)
               for filename_19_20 in glob.glob('../data/JC-*.csv')])
```

```python
# Print out 2019-2020 DataFrame
final_bike_data_df
```
